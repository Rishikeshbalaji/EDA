```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv('mtcars.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# heat map to plot null values in data base 
sns.heatmap(df.isnull())
```

```python
#let's check the distribution of hp(horse power) feature
sns.distplot(df['hp'],hist=True)
plt.show()
```

```python
df['hp'].mean()
```

```python
# hp feature is not highly skewed so we will impute missing values with mean

df['hp_mean']=df['hp'].fillna(df['hp'].mean())
df
```

```python
#outliers in hp
sns.boxplot(y = df.hp_mean)
```

```python
IQR=df['hp_mean'].quantile(0.75)-df['hp_mean'].quantile(0.25)
```

```python
#statistical way
lower_bridge=df['hp_mean'].quantile(0.25)-(IQR*1.5)
upper_bridge=df['hp_mean'].quantile(0.75)+(IQR*1.5)
print(lower_bridge)
print(upper_bridge)
```

```python
lower_bridge=df['hp_mean'].quantile(0.25)-(IQR*1.5)
upper_bridge=df['hp_mean'].quantile(0.95)+(IQR*1.5)
print(lower_bridge)
print(upper_bridge)
```
